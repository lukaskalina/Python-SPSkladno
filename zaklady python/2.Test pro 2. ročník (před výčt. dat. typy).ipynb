Vytvoř program, který požádá uživatele o zadání počtu bodů (0 - 100).

Na základě počtu bodů program přiřadí známku podle těchto pravidel:
* 90 až 100 bodů: "Výborně (1)"
* 75 až 89 bodů: "Chvalitebně (2)"
* 60 až 74 bodů: "Dobře (3)"
* 50 až 59 bodů: "Dostatečně (4)"
* 0 až 49 bodů: "Nedostatečně (5)"

Pokud uživatel zadá číslo mimo rozsah 0-100, program vypíše chybovou hlášku: "Neplatný počet bodů."

````python
vstup = int(input("zadejte počet bodů"))
print("počet bodů je:", vstup)

if vstup > 100:
    print("Neplatný")
elif vstup >= 90:
    print("výborný")
elif vstup >= 75:
    print("chvalitebné")
elif vstup >= 60:
    print("dobře")
elif vstup >= 50:
    print("dostatečné")
elif vstup >= 0:
    print("nedostatečné")
````
