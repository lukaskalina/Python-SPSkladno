**Vytvořte Třídu Room, která bude obsahovat atributy:**


1. název
2. seznamExitu
3. poklad
4. nepritel

**Vytvořte metody:**


1.   addExit, getNazev, getExity, getPoklad, getNepritel, toString
2.   setNazev





```python
class Room:
    def __init__(self, nazev ):
        self.nazev = nazev
        self.seznamExitu = []
        self.poklad = None
        self.nepritel = None
    def addExit(self , exit):
        self.seznamExitu.append(exit)
    def getNazev(self):
        return self.nazev
    def setNazev(self , nazev):
        self.nazev = nazev
    def getSeznamExitu(self):
        return [r.nazev for r in self.seznamExitu]
    def getPoklad(self):
        return self.poklad
    def getNepritel(self):
        return self.nepritel
    def Tostring(self):
        return str(self)
    def __str__(self):
        exit_names = [r.nazev for r in self.seznamExitu]
        return f"Room({self.nazev}) exity: {exit_names} poklad: {self.poklad} nepritel {self.nepritel}"

        
```

**Zde je vytvořená jednoduchá mapa s výpisy**

```python
room1 = Room("Vstup do jeskyně")
room2 = Room("Jezírko")
room3 = Room("Předsíň draka")
room4 = Room("Dračí slůj")
room5 = Room("Vyhlídka")

room1.addExit(room5)
room5.addExit(room1)
room1.addExit(room2)
room1.addExit(room4)
room2.addExit(room1)
room4.addExit(room1)
room4.addExit(room3)
room4.addExit(room5)
room3.addExit(room4)
room5.addExit(room4)

aktualni = room1
print(aktualni.getSeznamExitu())
print(aktualni.getNepritel())
print(aktualni.getPoklad())
print(aktualni)

#chci se přesunout tak si změním aktualní na jeden z východu
aktualni = aktualni.seznamExitu[0]
print(aktualni)
#aktualni = aktualni.seznamExitu[2] # již nemá 3 exit --> skončí chybou
#print(aktualni)
aktualni = aktualni.seznamExitu[1]
print(aktualni)
```
