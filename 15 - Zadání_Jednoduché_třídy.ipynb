Vytvořte metody, které budou moc získávat a měnit tyto atributy. Také vytvořte metodu, která vám vypíše vše tomto počasí


````python
from enum import Enum
class Pocasicko(Enum):
    SLUNECNO = ""
    ZATAZENO
    BOURKA
class Pocasi:
    def __init__(self, vitr, povet, teplota , vlhkost):
        self.vitr = vitr
        self.povet = povet
        self.teplota= teplota
        self.vlhkost = vlhkost
    def getvitr(self):
        return self.vitr
    def setvitr(self,vitr):
        self.vitr = vitr 
    def getpovet(self):
        return self.povet
    def setpovet(self,povet):
        self.povet = povet
    def getteplota(self):
        return self.teplota
    def setteplota(self,teplota):
        self.teplota = teplota
    def getvlhkost(self):
        return self.vlhkost
    def setvlhkost(self,vlhkost):
        self.vlhkost = vlhkost  
    def __str__(self):
        return """
        vitr: {} km/h
        teplota: {} °C
        vlhkost: {}  %
        počasí: {}
         """.format(self.vitr,self.povet,self.teplota,self.vlhkost)  
        
        
       
    
````
